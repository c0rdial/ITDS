## Filtering dataframe with condition (AND)

```python
((titanic.sex == 'male')& (titanic.age >=14))
```

```python
titanic[(titanic.sex =='male') & (titanic.age >=14)].loc[:,["survived", "pclass", "sex","age"]]
```

```python
titanic[(titanic.sex =='female') & (titanic.age < 14)].loc[:,["survived", "pclass", "sex","age"]]
```

```python
mask1=titanic.sex == 'male'
mask1.head()
```

```python
mask2 = titanic.age >= 14
mask2.head()
```

```python
(mask1 & mask2).head()
```

```python
male_adult = titanic.loc[mask1 & mask2, ["survived","pclass","sex","age"]]
male_adult.head(10)
```

## Filtering dataframe with conditions (OR)

```python
mask1 = titanic.sex == 'female'
mask1.head()
```

```python
mask2 = titanic.age < 14
mask2.head(10)
```

```python
titanic.loc[mask1 | mask2]
```

```python
wom_or_chi = titanic.loc[mask1 | mask2, ["survived", "pclass", "sex","age"]]
wom_or_chi.head()
```

```python
wom_or_chi.info()
```

```python
g1 = wom_or_chi.groupby(by="survived")
l1 = list(g1)
l1
```
